```
def separate(mix1):
  even1=[]
  odd1=[]
  for x in mix1:
    #print(x)
    if(x % 2 == 0):
      even1.append(x)
    else:
      odd1.append(x)

  return even1,odd1
```

```
def combine(list1,list2):
  l=[]
  l.extend(list1)
  l.extend(list2)
  l.sort()
  return l

#combine([4,5],[3,8])
```
