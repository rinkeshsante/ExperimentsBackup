```python
from tkinter import *

class calc():
    def __init__(self):
        self.gui = Tk()
        self.gui.title("Calculator")
        
        #gui.geometry('200x300')
        
        self.exp = ''
        self.eqtn = StringVar()
        
        self.display = Entry(self.gui , textvariable = self.eqtn)
        self.display.grid(columnspan= 4 ,ipadx = 50)
        
        self.n0 = Button(self.gui, text=' 0 ', command= lambda : self.charPress(0), height=1, width=7) 
        self.n0.grid(row=1, column=0) 
        
        self.n1 = Button(self.gui, text=' 1 ', command= lambda : self.charPress(1), height=1, width=7) 
        self.n1.grid(row=1, column=1) 
        
        self.n2 = Button(self.gui, text=' 2 ', command= lambda : self.charPress(2), height=1, width=7) 
        self.n2.grid(row=1, column=2) 
        
        self.n3 = Button(self.gui, text=' 3 ', command= lambda : self.charPress(3), height=1, width=7) 
        self.n3.grid(row=1, column=3) 
        
        self.n4 = Button(self.gui, text=' 4 ', command= lambda : self.charPress(4), height=1, width=7) 
        self.n4.grid(row=2, column=0) 
        
        self.n5 = Button(self.gui, text=' 5 ', command= lambda : self.charPress(5), height=1, width=7) 
        self.n5.grid(row=2, column=1) 
        
        self.n6 = Button(self.gui, text=' 6 ', command= lambda : self.charPress(6), height=1, width=7) 
        self.n6.grid(row=2, column=2) 
        
        self.n7 = Button(self.gui, text=' 7 ', command= lambda : self.charPress(7), height=1, width=7) 
        self.n7.grid(row=2, column=3)
        
        self.n8 = Button(self.gui, text=' 8 ', command= lambda : self.charPress(8), height=1, width=7) 
        self.n8.grid(row=3, column=1) 
        
        self.n9 = Button(self.gui, text=' 9 ', command= lambda : self.charPress(9), height=1, width=7) 
        self.n9.grid(row=3, column=2) 
        
        self.nClear = Button(self.gui, text=' C ', command= self.nClear, height=1, width=7) 
        self.nClear.grid(row=3, column=0)
        
        self.allClear = Button(self.gui, text=' AC ', command= self.allClear, height=1, width=7) 
        self.allClear.grid(row=3, column=3)
        
        self.nAdd = Button(self.gui, text=' + ', command= lambda : self.charPress('+'), height=1, width=7) 
        self.nAdd.grid(row=4, column=0) 
        
        self.nSub = Button(self.gui, text=' - ', command= lambda : self.charPress('-'), height=1, width=7) 
        self.nSub.grid(row=4, column=1)
        
        self.nMul = Button(self.gui, text=' * ', command= lambda : self.charPress('*'), height=1, width=7) 
        self.nMul.grid(row=4, column=2)
        
        self.nDiv = Button(self.gui, text=' / ', command= lambda : self.charPress('/'), height=1, width=7) 
        self.nDiv.grid(row=4, column=3)
        
        self.equal = Button(self.gui, text=' = ', command=self.evalValue, height=1, width=7) 
        self.equal.grid(row=5, column=3) 
        
        self.gui.mainloop()
    
    def charPress(self,char):
        self.exp = self.exp + str(char)
        self.eqtn.set(self.exp)

    def evalValue(self):
        try:
            self.exp = str(eval(self.exp))
        except:
            self.exp = 'Error'
        self.eqtn.set(self.exp)
        self.exp = ''
        
    def nClear(self):
        self.exp = self.exp[:-1]
        self.eqtn.set(self.exp)

    def allClear(self):
        self.exp = ''
        self.eqtn.set(self.exp)

y = calc()   
```

```python
# Registration Form

from tkinter import *
from tkinter import messagebox as mb


class registrationForm():
    def __init__(self):
        self.gui = Tk()
        self.gui.title('Registration Form')
        
       # self.gui.geometry('340x400')
        
        self.l1= Label(self.gui,text="Registration Form")
        self.l1.grid(columnspan= 4 ,ipadx=100) 

        self.l2= Label(self.gui,text="First Name :")
        self.l2.grid(row =1,column=0) 
        self.var1 = StringVar()
        self.Fname = Entry(self.gui , textvariable = self.var1)
        self.Fname.grid(row =1,column=1)
        
        self.l3= Label(self.gui,text="Last Name :")
        self.l3.grid(row =2,column=0) 
        self.var2 = StringVar()
        self.Lname= Entry(self.gui , textvariable = self.var2)
        self.Lname.grid(row =2,column=1)
        
        self.l4= Label(self.gui,text="Email :")
        self.l4.grid(row =3,column=0) 
        self.var3 = StringVar()
        self.mail = Entry(self.gui , textvariable = self.var3)
        self.mail.grid(row =3,column=1)
        
        self.l5= Label(self.gui,text="Mobile No :")
        self.l5.grid(row =4,column=0) 
        self.var4 = StringVar()
        self.nameD = Entry(self.gui , textvariable = self.var4)
        self.nameD.grid(row =4,column=1)
        
        
        self.l6= Label(self.gui,text="Gender :")
        self.l6.grid(row =6,column=0) 
        self.var = StringVar('')
        self.R1 = Radiobutton(self.gui, text="Male", variable=self.var, value='male')
        self.R1.grid(row =6,column=1)
        self.R2 = Radiobutton(self.gui, text="Female", variable=self.var, value='female')
        self.R2.grid(row =6,column=2)
        self.R3 = Radiobutton(self.gui, text="Other", variable=self.var, value='other')
        self.R3.grid(row =6,column=3)

        
        self.btn = Button(self.gui, text=' Submit ', command=self.Submit, height=1, width=7) 
        self.btn.grid(row=7, column=3) 
        
        self.gui.mainloop()
    
    def Submit(self):
        val = (self.var1.get(),self.var2.get(),self.var3.get(),self.var4.get(),self.var.get(),)
        mb.showinfo('data:',val)

        
        
x= registrationForm()

# this form print the submited value in the terminal
```
