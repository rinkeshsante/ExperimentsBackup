```python
import collections
# lists
lList= collections.deque()
```

```python
while True:
    msg = '''
    1: add right
    2: remove right
    3: add left
    4: remove left
    5: print all
    6: exit
'''
    ch = int(input(msg))
    
    if ch == 1:
        lList.append(input('num :'))
    elif ch == 2:
        print(lList.pop())
    elif ch == 3:
        lList.appendleft(input('num :'))
    elif ch == 4:
        print(lList.popleft())
    elif ch == 5:
        print('elements :')
        for item in lList:
            print(item)
    elif ch ==6:
        break
```

```python
# stack
stack = collections.deque()
```

```python
while True:
    msg = '''
    1: push()
    2: pop()
    3: print all
    4: exit
'''
    ch = int(input(msg))
    
    if ch == 1:
        stack.append(input('num :'))
    elif ch == 2:
        print(stack.pop())
    elif ch == 3:
        print('elements :')
        for item in stack:
            print(item)
    elif ch ==4:
        break
```
